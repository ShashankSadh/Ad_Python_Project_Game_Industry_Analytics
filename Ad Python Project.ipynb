```python
import pandas as pd

file_path = r"C:\Users\shash\Desktop\Projects\Python_Finale_Proj\Python_Finale_Proj_dataset.xlsx"

excel_data = pd.ExcelFile(file_path)
excel_data.sheet_names
```

```python
data = pd.read_excel(file_path, sheet_name="Sheet1")

data.head()
```

```python
data.info()

data.describe(include="all")
```

```python
data['Development_Year'] = data['Development_Year'].fillna(data['Development_Year'].median()).astype(int)
data['Release_Year'] = data['Release_Year'].fillna(data['Release_Year'].median()).astype(int)

# Replace missing categorical data with 'Unknown'
categorical_cols = ['Game_Title', 'Platform', 'Genre', 'Developer', 'Publisher', 'Country', 
                    'Online_Mode', 'In_Development', 'DLC_Available', 'Cross_Play']
data[categorical_cols] = data[categorical_cols].fillna('Unknown')

# Fill numerical columns with median values
numerical_cols = ['Player_Count', 'Rating', 'Price', 'Income', 'Worldwide_Ranking']
data[numerical_cols] = data[numerical_cols].fillna(data[numerical_cols].median())

# Convert boolean-like columns to categorical
boolean_cols = ['Online_Mode', 'In_Development', 'DLC_Available', 'Cross_Play']
data[boolean_cols] = data[boolean_cols].astype('category')

# Confirm cleaned dataset
data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# 1. Genre Popularity
plt.figure(figsize=(10, 6))
genre_counts = data['Genre'].value_counts()
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette="viridis")
plt.title("Genre Popularity", fontsize=16)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Games", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Revenue Trends by Year
plt.figure(figsize=(12, 6))
average_income_by_year = data.groupby('Release_Year')['Income'].mean()
sns.lineplot(x=average_income_by_year.index, y=average_income_by_year.values, marker="o", color="b")
plt.title("Average Revenue Trends by Release Year", fontsize=16)
plt.xlabel("Release Year", fontsize=12)
plt.ylabel("Average Revenue", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 3. Platform Distribution
plt.figure(figsize=(8, 6))
platform_counts = data['Platform'].value_counts()
platform_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title("Platform Distribution", fontsize=16)
plt.ylabel("")
plt.show()
```

```python
# 4. Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = data[['Player_Count', 'Rating', 'Price', 'Income', 'Worldwide_Ranking']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

```python
# 5. Scatter Plot: Rating vs. Income
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Rating'], y=data['Income'], hue=data['Platform'], palette="Set2", alpha=0.7)
plt.title("Rating vs. Income by Platform", fontsize=16)
plt.xlabel("Game Rating", fontsize=12)
plt.ylabel("Income", fontsize=12)
plt.legend(title="Platform", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 6. Games in Development by Year
plt.figure(figsize=(12, 6))
games_in_dev_by_year = data[data['In_Development'] == 'Yes'].groupby('Development_Year').size()
sns.lineplot(x=games_in_dev_by_year.index, y=games_in_dev_by_year.values, marker="o", color="purple")
plt.title("Games in Development by Year", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Games", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 7. Revenue by Country
plt.figure(figsize=(12, 6))
top_countries_revenue = data.groupby('Country')['Income'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_countries_revenue.values, y=top_countries_revenue.index, palette="coolwarm")
plt.title("Top Countries by Revenue", fontsize=16)
plt.xlabel("Total Revenue", fontsize=12)
plt.ylabel("Country", fontsize=12)
plt.show()
```

```python
# 8. Player Count by Genre
plt.figure(figsize=(12, 6))
sns.boxplot(x='Genre', y='Player_Count', data=data, palette="Set3")
plt.title("Player Count by Genre", fontsize=16)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Player Count", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. Average Rating by Genre
plt.figure(figsize=(12, 6))
avg_rating_by_genre = data.groupby('Genre')['Rating'].mean().sort_values(ascending=False)
sns.barplot(x=avg_rating_by_genre.values, y=avg_rating_by_genre.index, palette="viridis")
plt.title("Average Rating by Genre", fontsize=16)
plt.xlabel("Average Rating", fontsize=12)
plt.ylabel("Genre", fontsize=12)
plt.show()
```

```python
# 10. Revenue Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Income'], bins=20, kde=True, color="teal")
plt.title("Revenue Distribution", fontsize=16)
plt.xlabel("Income", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
# 11. Top Developers by Game Count
plt.figure(figsize=(12, 6))
top_developers = data['Developer'].value_counts().head(10)
sns.barplot(x=top_developers.values, y=top_developers.index, palette="mako")
plt.title("Top Developers by Game Count", fontsize=16)
plt.xlabel("Number of Games", fontsize=12)
plt.ylabel("Developer", fontsize=12)
plt.show()
```
